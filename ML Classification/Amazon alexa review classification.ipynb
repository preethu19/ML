## Data Cleaning/ Feature engineering

```python
dataset.drop(columns=['date','rating'], axis=1, inplace=True)
```

```python
dataset
```

```python
variation_dummies = pd.get_dummies(dataset['variation'], drop_first=True)
```

```python
variation_dummies
```

```python
dataset.drop(columns=['variation'], axis=1, inplace=True)
```

```python
dataset
```

```python
dataset = pd.concat([dataset, variation_dummies], axis=1)
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
dataset_count_vectorizer = vectorizer.fit_transform(dataset['verified_reviews'])
```

```python
dataset_count_vectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(dataset_count_vectorizer.toarray())
```

```python
dataset.drop(columns=['verified_reviews'], axis=1, inplace=True)
```

```python
dataset
```

```python
encoded_reviews = pd.DataFrame(dataset_count_vectorizer.toarray())
```

```python
dataset = pd.concat([dataset, encoded_reviews], axis=1)
```

```python
dataset
```

```python
X = dataset.drop('feedback', axis=1)
y = dataset['feedback']
```

```python
X.shape
```

```python
y.shape
```

## Model Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy')
```

```python
randomforest_classifier.fit(X_train, y_train)
```

## Model Testing

```python
y_predict_train = randomforest_classifier.predict(X_train)
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict = randomforest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

## Improve the Model

```python
dataset = pd.read_csv('amazon_alexa.tsv', sep='\t')
```

```python
dataset = pd.concat([dataset, pd.DataFrame(dataset_count_vectorizer.toarray())], axis=1)
```

```python
dataset
```

```python
dataset['length'] = dataset['verified_reviews'].apply(len)
```

```python
dataset
```

```python
X = dataset.drop(columns=['rating', 'date', 'variation', 'verified_reviews'], axis=1)
```

```python
y = dataset['feedback']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators=100, criterion='entropy')
randomforest_classifier.fit(X_train, y_train)
```

```python
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```
